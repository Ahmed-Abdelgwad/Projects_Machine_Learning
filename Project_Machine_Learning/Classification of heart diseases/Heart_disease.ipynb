```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import r2_score, confusion_matrix, accuracy_score, log_loss, classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LinearRegression, LogisticRegression
```

```python
le = LabelEncoder()
HeartDiseaseDataf['HeartDisease'] = le.fit_transform(HeartDiseaseDataf['HeartDisease'])
HeartDiseaseDataf['Smoking'] = le.fit_transform(HeartDiseaseDataf['Smoking'])
HeartDiseaseDataf['AlcoholDrinking'] = le.fit_transform(HeartDiseaseDataf['AlcoholDrinking'])
HeartDiseaseDataf['Stroke'] = le.fit_transform(HeartDiseaseDataf['Stroke'])
HeartDiseaseDataf['DiffWalking'] = le.fit_transform(HeartDiseaseDataf['DiffWalking'])
HeartDiseaseDataf['Sex'] = le.fit_transform(HeartDiseaseDataf['Sex'])
HeartDiseaseDataf['PhysicalActivity'] = le.fit_transform(HeartDiseaseDataf['PhysicalActivity'])
HeartDiseaseDataf['Asthma'] = le.fit_transform(HeartDiseaseDataf['Asthma'])
HeartDiseaseDataf['KidneyDisease'] = le.fit_transform(HeartDiseaseDataf['KidneyDisease'])
HeartDiseaseDataf['SkinCancer'] = le.fit_transform(HeartDiseaseDataf['SkinCancer'])
```

```python
df = pd.get_dummies(HeartDiseaseDataf, columns = ['AgeCategory', 'Race',  'Diabetic', 'GenHealth'])
```

```python
X = df.drop('HeartDisease', axis = 1)
y = df['HeartDisease']
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, test_size=0.3, random_state=100)
```

```python
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_val shape:", X_test.shape)
print("y_val shape:", y_test.shape)
```

```python
acc_baseline = y_train.value_counts(normalize=True).max()
print("Baseline Accuracy:", round(acc_baseline, 2))
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
train_acc = accuracy_score(y_train, model.predict(X_train))
test_acc = model.score(X_test, y_test)
print("Train Accuracy:", round(train_acc, 2))
print("Test Accuracy:", round(test_acc, 2))
```
